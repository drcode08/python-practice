1. What Are Escape Characters?

An escape character is a backslash (\) followed by another character.
It tells Python to treat the next character specially, not literally.

```python
print("He said, \"Hello!\"")
print("This is Divakar\\Reddy\" book")
# Here, \" is an escape sequence that allows quotes inside a string.
```

2. Common Escape Characters in Python



| Escape Sequence | Description     | Example                     | Output                     |
| --------------- | --------------- | --------------------------- | -------------------------- |
| `\\`            | Backslash       | `"This is a backslash: \\"` | `This is a backslash: \`   |
| `\'`            | Single quote    | `'It\'s Python'`            | `It's Python`              |
| `\"`            | Double quote    | `"He said, \"Hi\""`         | `He said, "Hi"`            |
| `\n`            | New line        | `"Hello\nWorld"`            | <pre>Hello<br>World</pre>  |
| `\t`            | Tab space       | `"A\tB\tC"`                 | `A    B    C`              |
| `\r`            | Carriage return | `"Hello\rWorld"`            | `World`                    |
| `\b`            | Backspace       | `"Helloo\b"`                | `Hello`                    |
| `\f`            | Form feed       | `"Hello\fWorld"`            | (page break between lines) |
| `\ooo`          | Octal value     | `"\110\145\154\154\157"`    | `Hello`                    |
| `\xhh`          | Hex value       | `"\x48\x65\x6C\x6C\x6F"`    | `Hello`                    |


3. Example: Using Escape Sequences

```python
print("Line1\nLine2")       # Newline
print("Column1\tColumn2")   # Tab
print("He said, \"Python is fun!\"")  # Quote
print("C:\\Users\\Divakar") # File path
```

4. Raw Strings (Ignore Escape Characters)
If you don’t want escape sequences to be processed, use a raw string by adding r or R before the quotes.

✅ Raw strings are especially useful for:

Windows file paths
Regular expressions

```python
path = r"C:\newfolder\test"
print(path)
```
