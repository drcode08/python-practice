### Class

A class is a blueprint used to create objects.

A class is a blueprint of a Car.

But the blueprint itself is NOT the car.

It tells how to build cars.

A Class Does NOT occupy memory

A class only becomes useful when we create objects from it.

```python
# Syntax This creates a class but no functionality yet.
class Person:
    pass

```

### What is an Object?

An object is an instance of a class.

If class = blueprint

then object = actual item created from that blueprint.

Examples:

class → Car blueprint

object → Tata, Audi, BMW, Tesla

class → Human blueprint

object → You, Me, Any person

Objects STORE DATA

### Objects can hold:

✔ attributes/properties (variables)

✔ methods (functions)

## How to Create an Object?

Here:

Person() creates a new object

p1 is a reference to that object

Object gets memory in the HEAP

p1 (the variable) lives in the STACK

```python
p1 = Person()
```

### Adding Attributes (Data) to a Class

We use a constructor to add data.

### Constructor (__init__)

A constructor is a special method that runs automatically when we create an object.

```python
# syntax

class ClassName:
    def __init__(self):
        # initialize data
        pass
```

```python
# Example

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Divakar", 30)
print(p1.name)
print(p1.age)
```

### Deep Explanation of Constructor

__init__ runs automatically.

self refers to the current object.

self.name means “store name inside the object”.

### What is self?

self refers to the object that is calling the method.

```python
# When you do:
p1 = Person("Alex", 22)

# Behind the scenes:

Person.__init__(p1, "Alex", 22)

# So:
# self = p1
# self.name = "Alex" stores data inside the object p1
```
