```python
import datetime

x = datetime.datetime.now()

print(x)

print(x.year)
print(x.month)
print(x.day)
print(x.hour)
```

#### Creating Date Objects

To create a date or datetime object in Python, you use the datetime() constructor from the datetime module.

##### Constructor format:

datetime(year, month, day, hour=0, minute=0, second=0, microsecond=0)

```python
from datetime import datetime

dt = datetime(2025,11,28,12,45,59)
print(dt)
```

#### Creating only a date object

```python
from datetime import date

d = date(2025, 11, 26)
print(d)
```

#### Getting current time

```python
from datetime import date

nowts = datetime.now()
print(nowts)
```

#### Creating time-only object

```python
from datetime import time

time = time(12,35,59)
print(time)
```

#### Summary

| Type     | Constructor  | Example                          |
| -------- | ------------ | -------------------------------- |
| Date     | `date()`     | `date(2025, 11, 26)`             |
| Time     | `time()`     | `time(14, 30)`                   |
| DateTime | `datetime()` | `datetime(2025, 11, 26, 14, 30)` |




#### Parsing Strings into datetime (string → datetime)

Use datetime.strptime:

```python
from datetime import datetime

dt = datetime.strptime("2025-11-27 12:45:59", "%Y-%m-%d %H:%M:%S")
print("formated time is..", dt)
```

#### Formatting datetime into Strings (datetime → string)

Use strftime:

```python
from datetime import datetime

now = datetime.now()
formatted = now.strftime("%Y-%m-%d %H:%M:%S")
print("formatted time is..", formatted)
```

#### Common formatting symbols

| Symbol | Meaning       | Example |
| ------ | ------------- | ------- |
| `%Y`   | Year          | 2025    |
| `%m`   | Month (01–12) | 11      |
| `%d`   | Day           | 26      |
| `%H`   | Hour (00–23)  | 14      |
| `%M`   | Minute        | 30      |
| `%S`   | Second        | 45      |




#### Adding or Subtracting Time (timedelta)

```python
from datetime import datetime, timedelta

now = datetime.now()
future = now + timedelta(days=5)
previous = now - timedelta(days=5)
past = now - timedelta(hours=3)

print(now)
print(future)
print(previous)
print(past)
```
