### Dictionaries in Python

A dictionary in Python is a collection of key–value pairs.

It is:

✔ Mutable (changeable)

✔ Unordered

✔ Indexed by keys, not by positions

✔ Doesn’t allow duplicate keys

✔ Dictionaries cannot have two items with the same key

```python
myDict = {"Name":"Divakar","City":"Bengaluru"}       
print(myDict)

print("Keys are...",myDict.keys())
print("Values are....",myDict.values())
print("Items in a Dict", myDict.items())
```

### 1. Creating Dictionaries

```python
# Using curly brackets
student = {"name": "John", "age": 20, "city": "Delhi"}
print(student)

# Using dict() constructor
employee = dict(id=101, name="Alice", salary=50000)
print(employee)

# Empty dictionary
d = {}
```

### 2. Accessing Values

get() is safer — it won’t cause an error if key doesn’t exist.

```python
print(student['name'])       #
print(student.get('age'))    #
```

### 3. Adding & Updating Items

```python
student['age'] = 30
student['college'] = 'JNTU'
print(student)
```

### 4. Removing Items

```python
print("Before pop", student)
student.pop("college")        # removes key 'city'
print("After pop", student)
student.popitem()          # removes last inserted pair
print("After pop_item....", student.keys())
student['grade'] = 'A'
print("student details...", student)
del student["grade"]         # delete specific key
print("After del...", student)
student.clear()            # empty the dictionary
print("after clear..", student)
```

### 5. Useful Dictionary Methods

```python
person = {"name": "Divakar", "age": 35}

print(person.keys())     # dict_keys(['name', 'age'])
print(person.values())   # dict_values(['Rahul', 25])
print(person.items())    # dict_items([('name', 'Rahul'), ('age', 25)])
```

### 6. Looping Through Dictionary

```python
for key in person:
    print(key,":", person[key])

# OR

for k,v in person.items():
    print(k,"=",v)
```

### 7. Nested Dictionary

```python
students = {
    101: {"name":"Divakar","marks": 800},
    102: {"name": "Vikram", "marks": 900}
}

print(students)
print(students[101]["name"])
print(students[102]["marks"])


for id in students:
    print(id)
   
    
```

### Copy Dictionary

```python
myDict = {"name":"Divakar","City":"Bengaluru"}

## Using copy()
newDict = myDict.copy()
print(type(newDict))
print(newDict)

## Using dict()

newDict1 = dict(myDict)
print(newDict1)

```

## Dictionay Methods

| Method           | Description                                                                                                 |
| ---------------- | ----------------------------------------------------------------------------------------------------------- |
| **clear()**      | Removes all the elements from the dictionary                                                                |
| **copy()**       | Returns a copy of the dictionary                                                                            |
| **fromkeys()**   | Returns a dictionary with the specified keys and value                                                      |
| **get()**        | Returns the value of the specified key                                                                      |
| **items()**      | Returns a list containing a tuple for each key–value pair                                                   |
| **keys()**       | Returns a list containing the dictionary's keys                                                             |
| **pop()**        | Removes the element with the specified key                                                                  |
| **popitem()**    | Removes the last inserted key–value pair                                                                    |
| **setdefault()** | Returns the value of the specified key; if the key does not exist, inserts the key with the specified value |
| **update()**     | Updates the dictionary with the specified key–value pairs                                                   |
| **values()**     | Returns a list of all the values in the dictionary                                                          |
