### 1. If Statement

Used to run a block of code only when a condition is true.

```python
# Sytax

# if condition:
#   #code

salary = 20000
if salary >= 18000:
    print("You are eligible for promotion")
```

### 2. If–Else Statement

Runs one block if true, another if false.

```python
salary = 20000

if salary >= 30000:
    print("You are eligible for promotion")
else:
    print("You are not eligible for promotion")
```

### 3. If–Elif–Else

Used for multiple conditions.

```python
score = int(input("Enter score: "))
print(score)

# score = 88

if score >= 95:
    print("Grade A")
elif score >= 75:
    print("Grade B")
elif score >= 50:
    print("Grade C")
else:
    print("Fail")
```

You will use if/elif/else to:

✔ control training loops

### Loops

In Python, we have two primitive loop commands:

- While Loop

- For Loop

** While loop runs until a condition becomes false.

```python
count = 0

while count <= 5:
    print(count)
    count += 1
```

```python
i = 0
name = 'DIVAKAR'
print("Length of the give string ", len(name))

while i < len(name):
    print(name[i])
    i += 1
```

```python
i = 1
while i < 6:
    print(i)
    i += 1
```

### 5. Loop Control Statements

#### - break:

The break statement immediately exits the nearest enclosing loop.

Execution continues with the first statement after the loop.

#### - continue:

The continue statement skips the rest of the current loop iteration.

The loop then moves on to the next iteration.

#### - pass

The pass statement does nothing.

It is used as a placeholder in situations where a statement is syntactically required but no action is needed (e.g., empty functions, loops, or conditionals).

```python
i = 1
while i < 10:
    print("Iteration...", i)
    i += 1
```

```python
i = 1
while i < 10:
    print("Iteration...", i)
    if i == 5:
        break
    i += 1
```

```python
i = 1
while i < 10:
    i += 1
    if i == 5:
        print("Skipped the Iteration for..", i)
        continue
    print("Iteration...", i)
```

```python
name = "Sachine_Tendulkar"
i = 0

while i < len(name):
    if name[i] == '_':
        i += 1
        continue
    print(name[i])
    i += 1
```

** A for loop is used for iterating over a sequence.

Sequecne can be list, tupe, set, string or a dictionary

```python
cities = ["bengaluru", "chennai", "bombay", "hyderabad"]
for city in cities:
    print(city)
```

```python
cities = ["bengaluru", "chennai", "bombay", "hyderabad"]
for city in cities:
    if city == 'bombay':
        break       # come out from the iteration
    print(city)
```

```python
cities = ["bengaluru", "chennai", "bombay", "hyderabad"]
for city in cities:
    if city == 'bombay':
        continue      # skips the current iteration
    print(city)
```

```python
cities = ["bengaluru", "chennai", "bombay", "hyderabad"]
for city in cities:
    if city == 'bombay':
        pass      # nothing does
    print(city)
```

### Nested Loop

A nested loop is a loop inside another loop.

The inner loop runs completely every time the outer loop runs one time.

```python
for x in range(3):  # Outer loop iterates for three times (0,1,2)
    for y in range(2): # Iterates for each outer loop value for (0,1) 2 times inner loop 
        print(x,y)
```

How it works the above nested loop

When x = 0, the inner loop runs with y = 0 and y = 1

When x = 1, the inner loop runs with y = 0 and y = 1

When x = 2, the inner loop runs with y = 0 and y = 1
