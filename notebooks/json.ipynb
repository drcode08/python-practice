#### convert JSON to python object

If you have a json string then you can parse it by using the json.loads() methos

```python
import json

x = '{"name":"Divakar", "city":"bengaluru"}'     # json string

y = json.loads(x)   # converted into dict
print(type(y))
print(y['name'])
print(y)

```

#### Convert Python Object to JSON

```python
pyObj = {"name":"Mike", "game":"football","year":"2025"}

x = json.dumps(pyObj)
print(type(x))
print(x)
```

```python
print(pyObj['name'])
```

```python
print(x['name'])
```

```python
import json

print(json.dumps({"name": "John", "age": 30}))
print(json.dumps(["apple", "bananas"]))
print(json.dumps(("apple", "bananas")))
print(json.dumps("hello"))
print(json.dumps(42))
print(json.dumps(31.76))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

#### Convert a Python object containing all the legal data types:

```python
import json

x = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}

print("Before json dump...", type(json.dumps(x)))
print(json.dumps(x))
print("After json dump...",type(x))
```

```python
## Indent will be added to each field
print(json.dumps(x, indent=4))
```

```python
## Keys gets sorted
print(json.dumps(x, indent=4, sort_keys=True))
```
