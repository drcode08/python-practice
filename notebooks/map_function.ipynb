### map() function

The map() function executes a specified function for each itel in an iterable. The item is sent to the functio as a parameter

map() is a built-in function that:

applies a function to each element of an iterable (like list, tuple, set) and returns a map object.

map(function, iterable)

    function → a function that will be applied to each element

    iterable → list, tuple, set, etc.

### Without map()

```python
myList = [1,2,3,4]
newList = []
for each in myList:
    newList.append(each*2)
print(newList)
```

### With map() Function

```python
myList = [1,2,3,4]

## Multilpy each item with the same number
def myFunc(num):
    return (num*num)

newList = list(map(myFunc, myList))
print(newList)

## Multiply each item with 2
result = map(lambda x: x * 2, myList)
print(list(result))
```

```python
lst1 = ['a','i','c','hi']
lst2 = ['m','s','at']


result = list(map(lambda x, y: x + y, lst1, lst2))
print(list(result)) ## 'hi' will be excluded as the lst doen't an corresponding element
```
