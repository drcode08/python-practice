### ðŸ§© 1. What is a List?

A list is an ordered, mutable (changeable) collection of items.

Lists can store different data types â€” numbers, strings, even other lists.

```python
my_list = ['Apple','MI','Samsung','Nokia']
print(my_list)
print(type(my_list))
```

### 2. Creating Lists

```python
empty_list = []
numbers = [1,2,3,4,5]
mixed_list = ["Diva", 1, 12.5, "Bengaluru"]
nested_list = [[1,2], [3], [4,5,7]]
print(empty_list)
print(numbers)
print(mixed_list)
print(nested_list)
```

### 3. Accessing List Items (Indexing)

Python lists are zero-indexed, meaning first element is index 0

Positive Index will take values from starting

Negative Index will take values from ending

We can specify range of indexes by specifying where to start and where to end

To check if specific items is prrsnt or not in list use "in" keyword

```python
fruits = ["apple", "banana", "oranges", "grapes"]
print(fruits[0])  # apple
print(fruits[1])  # banana
print(fruits[2])  # oranges
print(fruits[-1]) # grapes
```

### 4. Slicing Lists

```python
numbers = [10, 20, 30, 40, 50, 60]
print(numbers[1:4])   # [20, 30, 40]
print(numbers[:3])    # [10, 20, 30]
print(numbers[3:])    # [40, 50, 60]
print(numbers[::2])   # [10, 30, 50] (step = 2)
```

### 5. Modifying Lists

Lists are mutable â€” you can change, add, or remove elements.

To change a value of a specific item, refer to the index number

To change the range of list items, then refer to the range of the index where you want to insert values 

```python
fruits = ["India", "China", "Nepal"]
fruits.append("Russia")
print(fruits)
fruits[1] = "America"
print("Changed List", fruits)
```

```python
fruits = ["apple", "banana", "cherry"]
print(fruits)

fruits[1] = "mango"     # change value
print(fruits)
fruits.append("orange") # add element at end
print(fruits)
fruits.insert(1, "kiwi") # add at index 1

print(fruits)
```

### 6. Combining Lists

```python
list1 = [1,2,3]
list2 = [3,4,5]
combined = list1+list2
print(combined)
list1.extend(list2)
print(list1)
```

### 7. Removing Elements

| Method     | Description                             | Example                  |
| ---------- | --------------------------------------- | ------------------------ |
| `remove()` | Removes the first occurrence of a value | `fruits.remove("apple")` |
| `pop()`    | Removes by index and returns it         | `fruits.pop(1)`          |
| `del`      | Deletes by index or whole list          | `del fruits[0]`          |
| `clear()`  | Removes all items                       | `fruits.clear()`         |




```python
fruits = ["apple", "banana", "cherry","grapes"]
fruits.remove("grapes")
print("After remove", fruits)
fruits.append("mango")
print("Before pop method",fruits)
fruits.pop()
print("After pop method",fruits)
fruits.pop(1)    # pop() method will remove last item first, if we don't specify the index number
print("After pop(1)",fruits)  # ['apple', 'cherry']
del fruits[0]
print("After delete", fruits)
fruits.clear()
print("After clear", fruits)
```

### 9. Looping Through Lists

```python
cities = ['Bengaluru', 'Goa', 'Hyderabad', 'Chennai']

for city in cities:
    print(city)

print("##############")
for i in range(len(cities)):
    print(cities[i])
```

### 10. Useful List Functions

| Function   | Description    | Example           | Output                     |
| ---------- | -------------- | ----------------- | -------------------------- |
| `len()`    | Count elements | `len(fruits)`     | 3                          |
| `max()`    | Largest value  | `max(numbers)`    | 60                         |
| `min()`    | Smallest value | `min(numbers)`    | 10                         |
| `sum()`    | Sum of items   | `sum(numbers)`    | 210                        |
| `sorted()` | Sorts list     | `sorted(numbers)` | `[10, 20, 30, 40, 50, 60]` |




```python
mylist = [50,10,30,20,40]
fruits = ["apple","mango","grapes","banana","oranges"]
print(len(mylist))
print(max(mylist))
print(min(mylist))
print(sum(mylist))
print(sorted(mylist))
print(mylist.reverse())
fruits.sort()
print(fruits)
```

```python
fruits = ["apple","mango","grapes","banana","oranges"]
fruits.sort()  # Sort in ascending order
print(fruits)
```

```python
fruits = ["apple","mango","grapes","banana","oranges"]
fruits.reverse()  # Sort in descending order
print(fruits)
```

```python
fruits = ["apple","mango","grapes","banana","oranges"]
fruits.sort(reverse = True)   # Sort in descending order when reverse = True used 
print(fruits)
```

```python
mylist = ['apple','grapes','mango']
copy_list = mylist.copy()
print(copy_list)
```

### 11. List Comprehension (Shortcuts)

Quick way to create lists using loops in one line.

```python
squares = []
squares = [x**2 for x in range(6)]
print(squares)
```

##### With condition:

```python
even = [x for x in range(10) if x % 2 == 0]
print(even)  # [0, 2, 4, 6, 8]

odd = [y for y in range(10) if y % 2 != 0]
print(odd)
```

### 12. Nested Lists

Lists inside lists â€” for matrices or 2D structures.

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
print(matrix[1])
print(matrix[0])
print(matrix[0][0])
print(matrix[1][2])
print(matrix[2][0])
```
