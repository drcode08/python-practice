### File Handling

Python has several functions for creating, reading, updating, and deleting files.

The key function for working with files in Python is the open() function.

The open() function takes two parameters; filename, and mode.

There are four different methods (modes) for opening a file:



### Syntax

To open a file for reading it is enough to specify the name of the file:

f = open("demofile.txt")

f = open("demofile.txt", "rt")

Because "r" for read, and "t" for text are the default values, you do not need to specify them.

```python
myFile = open('<filepath>/demo.txt')

print(myFile.read())
```

```python
myFile1 = open('<filepath>/demo.txt', 'rt')

print(myFile1.read())
```

### Using the with statement

You can also use the with statement when opening a file:

Then you do not have to worry about closing your files, the with statement takes care of that.

```python
with open('<filepath>/demo.txt') as f:
    print(f.read())
```

### Close Files

It is a good practice to always close the file when you are done with it.

If you are not using the with statement, you must write a close statement in order to close the file:

```python
f = open('<filepath>/demo.txt', 'rt')
print(f.readline())  ## readline() method will read single line
f.close()
```

```python
f = open('<filepath>/demo.txt', 'rt')
for each in f:
    print(f.read())
f.close()
```

### Read Only Parts of the File

By default the read() method returns the whole text, but you can also specify how many characters you want to return:


```python
with open("<filepath>/demo.txt") as f:
    print(f.read(5)) ## It reads first 5 characters 
```

```python
with open("<filepath>/demo.txt") as f:
    print(f.read(1)) ## It reads first 5 characters 
```

### Write to an Existing File

To write to an existing file, you must add a parameter to the open() function:

"a" - Append - will append to the end of the file

"w" - Write - will overwrite any existing content

```python
# Open the file "demo.txt" and append content to the file:

f = open('<filepath>/demo.txt', 'a')
f.write("\napending the data to the existing data to the demo file")
```

```python
with open('<filepath>/demo.txt') as f:
    print(f.read())
```

### Overwrite Existing Content

To overwrite the existing content to the file, use the w parameter:

```python
#open and read the file before the overwriting:
with open("<filepath>/demo_overwrite.txt") as f:
  print(f.read())
```

```python
with open("<filepath>/demo_overwrite.txt","w") as f:
    f.write("This is new file for overwriting")

#open and read the file after the overwriting:
with open("<filepath>/demo_overwrite.txt") as f:
  print(f.read())
```

### Create a New File

To create a new file in Python, use the open() method, with one of the following parameters:

"x" - Create - will create a file, returns an error if the file exists

"a" - Append - will create a file if the specified file does not exists

"w" - Write - will create a file if the specified file does not exists


```python
f = open("<filepath>/newfile.txt",'x')
```
